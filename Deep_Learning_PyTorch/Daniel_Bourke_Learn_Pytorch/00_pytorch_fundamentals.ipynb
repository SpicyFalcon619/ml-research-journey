<div style="font-size: 15px;">

##### The main two rules for matrix multiplication:
1. The inner dimensions must match:
* `(3, 2) @ (3, 2)` won't work
* `(2, 3) @ (3, 2)` will work
* `(3, 2) @ (2, 3)` will work
2. The resulting matrix has the shape of the outer dimensions:
* `(2, 3) @ (3, 2)` -> `(2, 2)`
* `(3, 2) @ (2, 3)` -> `(3, 3)`

**Note: `@`** in Python is the symbol for matrix multiplication.

</div>

```python
# torch.rand(3, 2) @ torch.rand(3, 2) -> won't work because inner-dimension don't match
torch.rand(3, 2) @ torch.rand(2, 3)
```

### One of the most common errors in deep learning (shape errors)

```python
# Shapes need to be in the right way
tensor_A = torch.tensor([[1, 2],
                         [3, 4],
                         [5, 6]], dtype=torch.float32)

tensor_B = torch.tensor([[7, 10],
                         [8, 11], 
                         [9, 12]], dtype=torch.float32)
# torch.mm() is an alias for torch.matmul()
torch.matmul(tensor_A, tensor_B) # (this will show error)
```

```python
tensor_A.shape, tensor_B.shape
```

<div style="font-size: 15px;">

A **transpose** switches the dimensions of a given tensor
We can perform transposes in PyTorch using either:

1. `torch.transpose(input, dim0, dim1)` - where input is the desired tensor to transpose and dim0 and dim1 are the dimensions to be swapped.
2. `tensor.T` - where tensor is the desired tensor to transpose.

</div>

```python
# View tensor_A and tensor_B
print(tensor_A)
print(tensor_B)
```

```python
# View tensor_A and tensor_B.T
print(tensor_A)
print(tensor_B.T)
```

```python
import torch
from numpy import dtype
# The operation works when tensor_B is transposed
print(f"Original shapes: tensor_A = {tensor_A.shape}, tensor_B = {tensor_B.shape}\n")
print(f"New shapes: tensor_A = {tensor_A.shape} (same as above), tensor_B.T = {tensor_B.T.shape}\n")
print(f"Multiplying: {tensor_A.shape} * {tensor_B.T.shape} <- inner dimensions match\n")
print("Output:\n")
output = torch.matmul(tensor_A, tensor_B.T).type(torch.int32)
print(output)
print(f"\nOutput shape: {output.shape}")
```

## Reshaping, stacking, squeezing and unsqueezing tensors

<div style="font-size: 15px;">

* **Reshaping** - reshapes an input tensor to a defined shape
* **View** - Return a view of an input tensor of certain shape but keep the same memory as the original tensor
* **Stacking** - combine multiple tensors on top of each other (vstack) or side by side (hstack)
* **Squeeze** - removes all `1` dimensions from a tensor
* **Unsqueeze** - add a `1` dimension to a target tensor
* **Permute** - Return a view of the input with dimensions permuted (swapped) in a certain way

</div>

<div style="font-size: 15px;">

| Method | One-line description |
|---|---|
| `torch.reshape(input, shape)` | Reshapes `input` to `shape` (if compatible), can also use `torch.Tensor.reshape()`. |
| `Tensor.view(shape)` | Returns a view of the original tensor in a different `shape` but shares the same data as the original tensor. |
| `torch.stack(tensors, dim=0)` | Concatenates a sequence of `tensors` along a new dimension (`dim`), all `tensors` must be same size. |
| `torch.squeeze(input)` | Squeezes `input` to remove all the dimensions with value `1`. |
| `torch.unsqueeze(input, dim)` | Returns `input` with a dimension value of `1` added at `dim`. |
| `torch.permute(input, dims)` | Returns a *view* of the original `input` with its dimensions permuted (rearranged) to `dims`. |

</div>


```python
x = torch.arange(1., 10.)
x, x.shape, x.ndim
```

```python
# Add an extra dimension
x_reshaped_1 = x.reshape(1, 9)
x_reshaped_1, x_reshaped_1.shape, x_reshaped_1.ndim
```

```python
x_reshaped_2 = x.reshape(9, 1)
x_reshaped_2, x_reshaped_2.shape, x_reshaped_2.ndim
```

```python
x_reshaped_3 = x.reshape(3, 3)
x_reshaped_3, x_reshaped_3.shape, x_reshaped_3.ndim
```

```python
x = torch.arange(1., 11.)
x, x.shape, x.ndim
```

```python
x_reshaped = x.reshape(5, 2)
x_reshaped
```

```python
# Change the view
z = x.view(1, 10)
z, z.shape
# Changing the view of a tensor with torch.view() really only creates a new view of the same tensor. So changing the view changes the original tensor too
```

```python
# Changing z changes x
z[:, 0] = 5
z, x
```

```python
# Stack tensors on top of each other
x_stacked = torch.stack([x, x], dim=1) # try changing dim to dim=1 and see what happens
x_stacked
```

```python
# Remove extra dimension from x_reshaped
print(f"Previous tensor: {x_reshaped_1}")
print(f"Previous shape: {x_reshaped_1.shape}")

# Remove extra dimension from x_reshaped
x_squeezed = x_reshaped_1.squeeze()
print(f"\nNew tensor: {x_squeezed}")
print(f"New shape: {x_squeezed.shape}")
```

```python
# torch.unsqueeze() - adds a signle dimension to a target tensor at a specific dim
print(f"Previous target: {x_squeezed}")
print(f"Previous shape: {x_squeezed.shape}")

# Add an extra dimension with unsqueeze
x_unsqueezed = x_squeezed.unsqueeze(dim=0)
print(f"\nNew tensor: {x_unsqueezed}")
print(f"New shape: {x_unsqueezed.shape}")
```

```python
# torch.permute - rearranges the dimension of a target tensor in a specific order
x_original = torch.rand(size=(244, 244, 3))

# Permute the original tensor to rearrange the axis (or dim) order
x_permuted = x_original.permute(2, 0, 1) # shifts axis 0->1, 1->2, 2->0
print(f"Previous shape: {x_original.shape}")
print(f"New shape: {x_permuted.shape}")
```

## PyTorch tensors & NumPy

<div style="font-size: 15px;">

* Data in NumPy, want in PyTorch tensor -> `torch.from_numpy(ndarray)`
* PyTorch tensor to NumPy -> `torch.Tensor.numpy()`

</div>

```python
# NumPy array to tensor
array = np.arange(1.0, 8.0)
tensor = torch.from_numpy(array)
array, tensor
```

```python
# Changing the value of array

array = array + 1
array, tensor # Doesnt change the values of the tensor
```

```python
# Tensor to NumPy array

tensor = torch.ones(7)
numpy_tensor = tensor.numpy()
tensor, numpy_tensor
```

```python
# Changing the Tensor

tensor = tensor + 1
tensor, numpy_tensor # Doesnt change the values of the array
```

## Reproducibility (trying to take the random out of random)
<div style="font-size: 15px;">

In short:

`start with random numbers -> tensor operations -> try to make better (again and again and again)`

To reduce the randomness in neural networks and PyTorch comes the concept if **random seed**

Essentially what the random seed does ins "flavour" the randomness.

</div>

```python
random_tensor_A = torch.rand(3, 4)
random_tensor_B = torch.rand(3, 4)

print(random_tensor_A)
print(random_tensor_B)
print(random_tensor_A == random_tensor_B)
```

```python
# random but reproducible tensors

# Set the random seed
RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
random_tensor_C = torch.rand(3, 4)

RANDOM_SEED = 42
torch.manual_seed(RANDOM_SEED)
random_tensor_D = torch.rand(3, 4)

print(random_tensor_C)
print(random_tensor_D)
print(random_tensor_C == random_tensor_D)
```
